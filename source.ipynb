1 - Health Resources and Services Administration Mental Health Professional Shortage Area (HPSA). Gives data on mental health service shortages with geographic coverage. 

2 - CDC PLACES Project, which provides county- and city-level mental health indicators. 

3 - U.S. Census Bureau American Community Survey (ACS). This will be necessary for calculating the number of providers in the Cincinnati area over time.

## Approach and Analysis
These datasets can be related and merged using shared geographic identifiers such as county name and FIPS codes, as well as matching years. Provider counts and shortage designations from HRSA can be combined with population data from the ACS to calculate providers per 100,000 residents, while mental health distress indicators from the CDC PLACES Project can be aligned by year and county to compare trends in access and community need over time.
